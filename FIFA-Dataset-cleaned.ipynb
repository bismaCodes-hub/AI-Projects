```python
import pandas as pd
import re
import numpy as np
```

```python
fifa_data=pd.read_csv("fifa21 raw data v2.csv", low_memory=False)
```

```python
cols=fifa_data.columns
```

```python
fifa_data["Hits"]=pd.to_numeric(fifa_data["Hits"],errors="coerce")
```

```python
fifa_data["Main Position"]=fifa_data["Positions"].str.split(',').str[0].str.strip()
```

```python
fifa_data["Hits"]=fifa_data.groupby("Main Position")["Hits"].transform(lambda x:x.fillna(fifa_data["Hits"].mean()) )
```

```python
fifa_data.rename(columns={"Height": "Height_cm"}, inplace=True)
```

```python
def convert_height(value):
 if isinstance(value,str):
    value=value.strip()
    if "cm" in value:
        return float(value.replace("cm",'').strip())
    match=re.match(r"^(\d+)'\s*(\d+)",value) 
    if match:
        feet = int(match.group(1))*30.48
        inches = int(match.group(2))*2.54
        return round((inches + feet),1)
```

```python
fifa_data["Height_cm"]=fifa_data["Height_cm"].apply(convert_height)
```

```python
fifa_data.rename(columns={"Weight":"Weight_kg"}, inplace=True)
```

```python
def convert_weight(val):
  if isinstance(val,str):
      val=val.strip()
      if "kg" in val:
          return float(val.replace("kg","").strip())
      elif "lbs" in val:
          val=float(val.replace("lbs","").strip())
          return round((val*0.453592),1)
```

```python
fifa_data["Weight_kg"]=fifa_data["Weight_kg"].apply(convert_weight)
```

```python
def convert_wage(money):
    if isinstance(money,str):
        money=money.strip()
        if ("€" in money) and ("K" in money):
           money = float(money.replace("€","").replace("K","").strip())
           return money*1000
        elif ("€" in money) and ("M" in money):
            money = float(money.replace("€","").replace("M","").strip())
            return money*1000000
        elif "K" in money:
            money = float(money.replace("K","").strip())
            return money*1000
        elif "M" in money:
            money = float(money.replace("M","").strip())
            return money*1000000
        elif "€" in money:
            return float(money.replace("€","").strip())
```

```python
fifa_data["Wage"]=fifa_data["Wage"].apply(convert_wage)
```

```python
fifa_data["Value"]=fifa_data["Value"].apply(convert_wage)
```

```python
fifa_data["Release Clause"]=fifa_data["Release Clause"].apply(convert_wage)
```

```python
fifa_data["W/F"].unique()
```

```python
def convert_star(star):
    if isinstance(star,str):
        star = star.replace("★", "").strip()
        return pd.to_numeric(star.replace("★","").strip())
    else:
        return star
```

```python
fifa_data["W/F"]=fifa_data["W/F"].apply(convert_star)
```

```python
fifa_data["SM"]=fifa_data["SM"].apply(convert_star)
```

```python
fifa_data["IR"]=fifa_data["IR"].apply(convert_star)
```

```python
fifa_data["Loan Date End"]=fifa_data["Loan Date End"].fillna("Not Loaned")
```

```python
fifa_data["Joined"] = pd.to_datetime(fifa_data["Joined"], format="%b %d, %Y")
```

```python
def convert_contract(time):
    if isinstance(time,str):
        time=time.strip()
        match0=re.match(r"^(\d{4})\s*~\s*(\d{4})&",time)
        if match0:
            start_year=int(match0.group(1))
            end_year=int(match0.group(2))
            return (start_year,end_year)

        elif "On Loan" in time:
            time=time.replace("On Loan","").strip()
            return pd.to_datetime(time)
        elif time.lower() == "free":
            return time.replace("free",np.nan)
        elif "-" in time:           
            return time.replace("-",np.nan)
```

```python
fifa_data["Contract"]=fifa_data["Contract"].apply(convert_contract)
```
